```python
from typing import Tuple
import math
import matplotlib.pyplot as plt
import generic
plt.rcParams['figure.figsize'] = [30, 10]
plt.grid()
# moving averager from Table 1
# A0 set below unity to keep network stable
ma = generic.SAKNetwork(
    A=[0.8, 1, 0], 
    B=[1/8, 0, 0],
    C=1,
    N=8,
)


response = ma.get_frequency_response(
    freq_range=[0, 4],
    sample_rate=4,
    num_steps=666,
    logy=True
)

# calculate again but with scipy
import scipy.signal as sig

plt.subplot(121, ylim=(-50, 20))
plt.grid()
plt.title('Magnitude (dB)')
plt.scatter(response[0], response[1], marker='.')
plt.subplot(122, ylim=(-5, 5))
plt.grid()
plt.title('Phase (pi radians)')
plt.scatter(response[0], response[2], marker='.')
```

```python
# integrator from Table 1
itg = generic.SAKNetwork(
    A=[0.5, 1, 0],
    B=[1, 0, 0],
    C=0,
    N=1,
)

response = itg.get_frequency_response(
    freq_range=[0, 1],
    sample_rate=1,
    num_steps=1000
)
plt.subplot(121, ylim=(-20, 0))
plt.grid()
plt.title('Magnitude')
plt.scatter(response[0], response[1], marker='.')
plt.subplot(122, ylim=(-2, 2))
plt.grid()
plt.title('Phase')
plt.scatter(response[0], response[2], marker='.')
```

```python
# audio comb from Table 2
itg = generic.SAKNetwork(
    A=[0.825, 0, 0.2],
    B=[1, 0, 0],
    C=0,
    N=0,
)

response = itg.get_frequency_response(
    freq_range=[0, 1],
    sample_rate=1,
    num_steps=1000
)
plt.subplot(121, ylim=(-4, 0))
plt.grid()
plt.title('Magnitude')
plt.scatter(response[0], response[1], marker='.')
plt.subplot(122, ylim=(-.5, .5))
plt.grid()
plt.title('Phase')
plt.scatter(response[0], response[2], marker='.')
```

```python
# comb filter
itg = generic.SAKNetwork(
    A=[0.7, 0, 0],
    B=[0.7, 0, 0],
    C=1,
    N=128,
)

response = itg.get_frequency_response(
    freq_range=[0, 128e3],
    sample_rate=128e3,
    num_steps=64000
)
plt.subplot(121)
plt.grid()
plt.title('Magnitude')
plt.scatter(response[0][0:10000], response[1][0:10000], marker='.')
plt.subplot(122)
plt.grid()
plt.title('Phase')
plt.scatter(response[0], response[2], marker='.')
```

```python
# first order IIR (non-SAK version)

sample_rate = 1e3

# 10Hz cutoff, 100Hz sampling freq
lpf = generic.FirstOrderIIRLowPass(fc=100, fs=sample_rate)
response = lpf.get_frequency_response(
    freq_range=[0, 500],
    sample_rate=sample_rate,
    num_steps=1000
)


plt.subplot(121)
plt.grid()
plt.title('Magnitude')
plt.scatter(response[0], response[1], marker='.')
plt.subplot(122)
plt.grid()
plt.title('Phase')
plt.scatter(response[0], response[2], marker='.')
```
